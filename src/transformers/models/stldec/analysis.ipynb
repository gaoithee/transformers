```python
import pandas as pd
from utils import exact_match, cosine_similarity, euclidean_distance
import matplotlib.pyplot as plt
```

```python
s_7200 = pd.read_csv('formulae_mining/step_7200_formulae.csv')
s_8000 = pd.read_csv('formulae_mining/step_8000_formulae.csv')
s_9600 = pd.read_csv('formulae_mining/step_9600_formulae.csv')
s_10000 = pd.read_csv('formulae_mining/step_10000_formulae.csv')
s_10400 = pd.read_csv('formulae_mining/step_10400_formulae.csv')
# s_11600 = pd.read_csv('formulae_mining/step_11600_formulae.csv')
s_12000 = pd.read_csv('formulae_mining/step_12000_formulae.csv')
s_15000 = pd.read_csv('formulae_mining/step_15000_formulae.csv')
s_16000 = pd.read_csv('formulae_mining/step_16000_formulae.csv')
s_17500 = pd.read_csv('formulae_mining/step_17500_formulae.csv')
# s_19500 = pd.read_csv('formulae_mining/step_19500_formulae.csv')
s_20000 = pd.read_csv('formulae_mining/step_20000_formulae.csv')
# s_22500 = pd.read_csv('formulae_mining/step_22500_formulae.csv')
s_23500 = pd.read_csv('formulae_mining/step_23500_formulae.csv')
s_28000 = pd.read_csv('formulae_mining/step_28000_formulae.csv')
s_29000 = pd.read_csv('formulae_mining/step_29000_formulae.csv')
s_30000 = pd.read_csv('formulae_mining/step_30000_formulae.csv')
s_31500 = pd.read_csv('formulae_mining/step_31500_formulae.csv')
s_32000 = pd.read_csv('formulae_mining/step_32000_formulae.csv')
s_32500 = pd.read_csv('formulae_mining/step_32500_formulae.csv')
```

```python
steps = [7200, 8000, 9600, 10000, 10400, 12000, 15000, 16000, 17500, 20000, 23500, 28000, 29000, 30000, 31500, 32000, 32500]
datasets = [s_7200, s_8000, s_9600, s_10000, s_10400, s_12000, s_15000, s_16000, s_17500, s_20000, s_23500, s_28000, s_29000, s_30000, s_31500, s_32000, s_32500]

minima = []
maxima = []
means = []

for i in datasets:
    minima.append(euclidean_distance(i)[0])
    maxima.append(euclidean_distance(i)[2])
    means.append(euclidean_distance(i)[1])
```

```python
# Creare il grafico
plt.figure(figsize=(10, 6))

# Traccia le linee per min, mean, max
plt.plot(steps, minima, label='Min', color='blue', marker='o')
plt.plot(steps, means, label='Mean', color='green', marker='s')
plt.plot(steps, maxima, label='Max', color='red', marker='^')

# Etichette degli assi
plt.xlabel('Steps')
plt.ylabel('Euclidean distance')

# Aggiungi il titolo e la legenda
plt.title('Min, mean and max euclidean distance (between generated and gold) across steps')
plt.legend()

# Mostra il grafico
plt.grid(True)
plt.show()
```

```python
minima_cos = []
maxima_cos = []
means_cos = []

for i in datasets:
    minima_cos.append(cosine_similarity(i)[0])
    maxima_cos.append(cosine_similarity(i)[2])
    means_cos.append(cosine_similarity(i)[1])
```

```python
# Creare il grafico
plt.figure(figsize=(10, 6))

# Traccia le linee per min, mean, max
plt.plot(steps, minima_cos, label='Min', color='blue', marker='o')
plt.plot(steps, means_cos, label='Mean', color='green', marker='s')
plt.plot(steps, maxima_cos, label='Max', color='red', marker='^')

# Etichette degli assi
plt.xlabel('Steps')
plt.ylabel('Cosine similarity')

# Aggiungi il titolo e la legenda
plt.title('Min, mean and max cosine similarity (between generated and gold) across steps')
plt.legend()

# Mostra il grafico
plt.grid(True)
plt.show()
```

```python
exact = []
for i in datasets:
    exact.append(exact_match(i)[0])
```

```python
# Creare il grafico
plt.figure(figsize=(10, 6))

# Traccia le linee per min, mean, max
plt.plot(steps, exact, label='Min', color='blue')

# Etichette degli assi
plt.xlabel('Steps')
plt.ylabel('Exact match percentage')

# Aggiungi il titolo e la legenda
plt.title('Average exact match percentage (between generated and gold) across steps')
plt.legend()

# Mostra il grafico
plt.grid(True)
plt.show()
```
